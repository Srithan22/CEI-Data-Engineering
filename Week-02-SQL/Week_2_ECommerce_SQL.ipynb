Aggregation 

```sql
SELECT COUNT(*) AS total_orders
FROM orders;
```

```sql
SELECT SUM(total_amount) AS total_revenue
FROM orders
WHERE status='Delivered';
```

```sql
SELECT category,
       AVG(unit_price) AS avg_price
FROM products
GROUP BY category;
```

```sql
SELECT status,
       COUNT(*) AS order_count,
       SUM(total_amount) AS total_revenue
FROM orders
GROUP BY status
ORDER BY total_revenue DESC;
```

```sql
SELECT category,
       MAX(unit_price) AS max_price,
       MIN(unit_price) AS min_price
FROM products
GROUP BY category;
```

```sql
SELECT category,
       AVG(unit_price) AS avg_price
FROM products
GROUP BY category
HAVING AVG(unit_price) > 2000;
```

Joins & Relationships

```sql
SELECT o.order_id,
       o.order_date,
       c.first_name,
       c.last_name,
       o.total_amount
FROM orders o
INNER JOIN customers c
ON o.customer_id = c.customer_id;
```

```sql
SELECT c.customer_id,
       c.first_name,
       o.order_id,
       o.order_date
FROM customers c
LEFT JOIN orders o
ON c.customer_id = o.customer_id;
```

```sql
SELECT o.order_id,
       p.product_name,
       oi.quantity,
       oi.unit_price,
       oi.discount_pct
FROM orders o
JOIN order_items oi
ON o.order_id = oi.order_id
JOIN products p
ON oi.product_id = p.product_id;
```

LEFT JOIN:
Returns all rows from left table and matching rows from right table.
RIGHT JOIN:
Returns all rows from right table and matching rows from left table.
FULL OUTER JOIN:
Returns all records from both tables whether matched or not.

Foreign Keys:

orders.customer_id -> customers.customer_id
order_items.order_id -> orders.order_id
order_items.product_id -> products.product_id
If customer_id=999 does not exist, insertion fails due to Foreign Key constraint violation.

Advanced Concepts (CASE, ACID, Transactions)

```sql
SELECT product_name,
       unit_price,
       CASE
           WHEN unit_price < 1000 THEN 'Budget'
           WHEN unit_price BETWEEN 1000 AND 3000 THEN 'Mid-Range'
           ELSE 'Premium'
       END AS price_tier
FROM products;
```

```sql
SELECT
SUM(CASE WHEN status='Delivered' THEN 1 ELSE 0 END) AS Delivered,
SUM(CASE WHEN status<>'Delivered' THEN 1 ELSE 0 END) AS Not_Delivered
FROM orders;
```

A - Atomicity
All operations succeed or none.
C - Consistency
Database remains valid before and after transaction.
I - Isolation
Transactions do not interfere with each other.
D - Durability
Committed data remains saved permanently.
Example:
Bank Transfer

Atomicity:
Money deducted and credited together.
Consistency:
Total balance remains correct.
Isolation:
Two transfers at same time do not affect each other.
Durability:
After commit, transfer remains saved even after system crash.
