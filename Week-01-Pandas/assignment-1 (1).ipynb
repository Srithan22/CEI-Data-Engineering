Dataset: (Shopping Dataset)
The dataset is distributed across multiple category-wise CSV files.
For easier analysis, all files will first be combined into a single dataset.


```python
import pandas as pd
import os
```

Only CSV files are required for this assignment, so I filtered them separately.

```python
folder_path = "/Workspace/Users/srithankodumuru22@gmail.com/DbAcademy/shopping_data"
all_f = os.listdir(folder_path)
csv_f= []

for file_name in all_f:
    if file_name.endswith(".csv"):
        csv_f.append(file_name)

print("CSV files found :", len(csv_f))
```

I combined all CSV files into a single dataframe.

```python
all_products = []

for file_name in csv_f:
    current_path = os.path.join(folder_path,file_name)
    current_frame = pd.read_csv(current_path)
    current_frame["source_category"] = file_name.replace(".csv", "")
    all_products.append(current_frame)

shopping_data = pd.concat(all_products,ignore_index=True)
print("Combined dataset shape:", shopping_data.shape)
```

Checking for missing values before performing operations

```python
shopping_data.isnull().sum().sort_values(ascending=False)
```

Most missing values were found in only a few columns.
For numerical columns, I replaced missing values using the median.

```python
numeric_columns = shopping_data.select_dtypes(include="number").columns

for column in numeric_columns:
    shopping_data[column] = (
        shopping_data[column].fillna(shopping_data[column].median())
    )
```

Handling duplicate records 

```python
count = shopping_data.duplicated().sum()
print("Duplicate rows found:", count)

if count > 0:
    shopping_data = shopping_data.drop_duplicates()
    print("Duplicates removed")
else:
    print("No duplicate records were found")
```

Creating a new feature called popularity_score.

This feature combines product rating and the number of ratings received. A product with a high rating and many reviews is generally considered more reliable.

```python
shopping_data["popularity_score"] = (
    shopping_data["rating"] * shopping_data["ratings_count"]
)
```

The popularity_score makes it easy to spot products that not only have high ratings, but are also trusted by many customers based on the number of reviews.

```python
output_path = ("/Workspace/Users/srithankodumuru22@gmail.com/DbAcademy/cleaned_shopping_dataset.csv")

shopping_data.to_csv(output_path, index=False)
print("Dataset saved successfully")
```
